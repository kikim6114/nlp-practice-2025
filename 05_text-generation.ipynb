## 5.2 Greedy Search Decoding

- ì—°ì†ì ì¸ ëª¨ë¸ ì¶œë ¥ì—ì„œ ì´ì‚°ì ì¸ í† í°ì„ ì–»ëŠ” ê°€ì¥ ê°„ë‹¨í•œ ë””ì½”ë”© ë°©ë²•ì€ ê° íƒ€ì„ìŠ¤í…ì—ì„œ ìµœëŒ€ í™•ë¥ ì„ ê°–ëŠ” í† í°ì„ íƒìš•ì ìœ¼ë¡œ ì„ íƒí•˜ëŠ” ê²ƒ.
- GPT2-XL ë¡œë“œ(íŒŒë¼ë¯¸í„° ìˆ˜ = 1.5B)

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"device = {device}")
# ì½”ë©ì˜ ê²½ìš° gpt2-xlì„ ì‚¬ìš©í•˜ë©´ ë©”ëª¨ë¦¬ ë¶€ì¡± ì—ëŸ¬ê°€ ë°œìƒí•©ë‹ˆë‹¤.
# ëŒ€ì‹  "gpt" ë˜ëŠ” "gpt2-large"ë¡œ ì§€ì •í•˜ê±°ë‚˜ ì½”ë© í”„ë¡œë¥¼ ì‚¬ìš©í•˜ì„¸ìš”.
model_name = "gpt2-xl"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name).to(device)
```

- ğŸ¤—íŠ¸ëœìŠ¤í¬ë¨¸ìŠ¤ëŠ” GPT-2 ê°™ì€ ìê·€íšŒê·€(autoregressive) ëª¨ë¸ì„ ìœ„í•´ `generate()` í•¨ìˆ˜ë¥¼ ì œê³µí•˜ì§€ë§Œ ì‘ë™ ë°©ì‹ì„ ì´í•´í•˜ê¸° ìœ„í•´ ë””ì½”ë”© ë©”ì„œë“œë¥¼ ì§ì ‘ êµ¬í˜„í•´ ë³¸ë‹¤.
- ê° íƒ€ì„ìŠ¤í…ë§ˆë‹¤ ê°€ì¥ ë†’ì€ í™•ë¥ ì„ ê°–ëŠ” í† í° 5ê°œë¥¼ ì €ì¥í•œë‹¤(beam searchë¥¼ ìœ„í•´ì„œê°€ ì•„ë‹ˆë¼ ì‹œê°ì  í™•ì¸ì„ ìœ„í•œ ë‹¨ìˆœ ì €ì¥).

```python
import pandas as pd

input_txt = "Transformers are the"
input_ids = tokenizer(input_txt, return_tensors="pt")["input_ids"].to(device)  # "pt":íŒŒì´í† ì¹˜ í…ì„œ
iterations = []
n_steps = 8  # ë””ì½”ë”© ë°˜ë³µ ìˆ˜
choices_per_step = 5 

with torch.no_grad():
    for _ in range(n_steps):
        iteration = dict()
        iteration["Input"] = tokenizer.decode(input_ids[0])
        output = model(input_ids=input_ids)
        # ì²« ë²ˆì§¸ ë°°ì¹˜ì˜ ë§ˆì§€ë§‰ í† í°ì˜ ë¡œì§“ì„ ì„ íƒí•´ ì†Œí”„íŠ¸ë§¥ìŠ¤ë¥¼ ì ìš©í•©ë‹ˆë‹¤.
        next_token_logits = output.logits[0, -1, :]
        next_token_probs = torch.softmax(next_token_logits, dim=-1)
        sorted_ids = torch.argsort(next_token_probs, dim=-1, descending=True)
        # ê°€ì¥ ë†’ì€ í™•ë¥ ì˜ í† í°ì„ ì €ì¥í•©ë‹ˆë‹¤.
        for choice_idx in range(choices_per_step):
            token_id = sorted_ids[choice_idx]
            token_prob = next_token_probs[token_id].cpu().numpy()
            token_choice = (
                f"{tokenizer.decode(token_id)} ({100 * token_prob:.2f}%)"
            )
            iteration[f"Choice {choice_idx+1}"] = token_choice
        # ì˜ˆì¸¡í•œ ë‹¤ìŒ í† í°ì„ ì…ë ¥ì— ì¶”ê°€í•©ë‹ˆë‹¤.
        input_ids = torch.cat([input_ids, sorted_ids[None, 0, None]], dim=-1)
        iterations.append(iteration)

pd.DataFrame(iterations)
```

- ë” ë³µì¡í•œ ë””ì½”ë”© ë°©ë²•ì„ ìœ„í•´ ğŸ¤—íŠ¸ëœìŠ¤í¬ë¨¸ìŠ¤ì— ë‚´ì¥ëœ `generate()` í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•œë‹¤.
- `do_sample=True`:
  - multinomial sampling, beam-search multinomial sampling, Top-K sampling, Top-p sampling
- `do_sample=False`:
  - beam search decoding(ì´ ë•ŒëŠ” `num_beams=1` ì´ë©´ greedy decoding). 

```python
input_ids = tokenizer(input_txt, return_tensors="pt")["input_ids"].to(device)
output = model.generate(input_ids, max_new_tokens=n_steps, do_sample=False)
print(tokenizer.decode(output[0]))
```

- OpenAI ìœ ë‹ˆì½˜ ê¸°ì‚¬ë¥¼ ì¬í˜„í•´ë³´ì.
- ê¸´ ë¬¸ì¥ì„ ìƒì„±í•  ìˆ˜ ìˆê²Œ `max_length=128`ë¡œ í• ë‹¹

```python
max_length = 128
input_txt = """In a shocking finding, scientist discovered \
a herd of unicorns living in a remote, previously unexplored \
valley, in the Andes Mountains. Even more surprising to the \
researchers was the fact that the unicorns spoke perfect English.\n\n
"""
input_ids = tokenizer(input_txt, return_tensors="pt")["input_ids"].to(device)
output_greedy = model.generate(input_ids, max_length=max_length,
                               do_sample=False)
print(tokenizer.decode(output_greedy[0]))
```

- ì•ì—ì„œ ë³¸ ê¸°ì‚¬ì™€ ë‹¤ë¥´ë‹¤!!
- ê°™ì€ ë¬¸ì¥ì´ ë°˜ë³µë˜ê³  ìˆë‹¤ --> greedy decodingì˜ ì¹˜ëª…ì  ë‹¨ì .  --> **Beam Search Decodingì„ ì“°ì!**

## 5.3 ë¹” ì„œì¹˜ ë””ì½”ë”©

- ë¡œê·¸í™•ë¥ ì„ ì‚¬ìš©í•˜ëŠ” ì´ìœ ëŠ”?  --> ê³„ì†ëœ ê³±ì…ˆìœ¼ë¡œ  underflowê°€ ë°œìƒí•˜ëŠ” ê²ƒì„ ë°©ì§€í•˜ê¸° ìœ„í•´
- ìƒì„±í•  ì‹œí€€ìŠ¤ ê¸¸ì´ê°€ 1024ì´ê³  ê° í† í°ì˜ í™•ë¥ ì´ 0.5ë¼ë©´,

- model()ì—ì„œ ë°›ì€ ì •ê·œí™”ë˜ì§€ ì•Šì€ ë¡œì§“ outputì„ log_softmaxë¡œ ì „ì²´ ì–´íœ˜ì— ëŒ€í•´ í™•ë¥ ë¶„í¬ logpë¥¼ ë§Œë“ ë‹¤
- logpì—ì„œ ì‹œí€€ìŠ¤ì— ìˆëŠ” í† í°ì˜ labelsì— í•´ë‹¹í•˜ëŠ” ë¡œê·¸í™•ë¥ ê°’ë“¤ë§Œ gatherí•´ì„œ logp_labelì— ë‹´ëŠ”ë‹¤ 

```python
import torch.nn.functional as F

def log_probs_from_logits(logits, labels):
    logp = F.log_softmax(logits, dim=-1)
    logp_label = torch.gather(logp, 2, labels.unsqueeze(2)).squeeze(-1)
    return logp_label
```

```python
def sequence_logprob(model, labels, input_len=0):
    with torch.no_grad():
        output = model(labels)
        log_probs = log_probs_from_logits(
            output.logits[:, :-1, :], labels[:, 1:])
        seq_log_prob = torch.sum(log_probs[:, input_len:])  # ì‹œí€€ìŠ¤ ì „ì²´ì— ëŒ€í•œ ë¡œê·¸í™•ë¥ 
    return seq_log_prob.cpu().numpy()
```

- logitsê³¼ labelsì˜ ì°¨ì›ì´ ì¼ì¹˜í•˜ë„ë¡ ì •ë ¬í•˜ëŠ” ê²ƒì´ ì¤‘ìš”.
- ëª¨ë¸ì€ ë‹¤ìŒ í† í°ì„ ì˜ˆì¸¡í•˜ë¯€ë¡œ, ì²«ë²ˆ labelì— ëŒ€í•´ì„œëŠ” logitì„ êµ¬í•˜ì§€ ì•Šìœ¼ë©°,
- ë§ˆì§€ë§‰ logitì€ ì´ì— ëŒ€í•œ ground truth í† í°ì´ ì—†ê¸° ë•Œë¬¸ì— í•„ìš”ì¹˜ ì•Šë‹¤.

OpenAI í”„ë¡¬í”„íŠ¸ì— ëŒ€í•œ Greedy Decoderì˜ ì‹œí€€ìŠ¤ ë¡œê·¸ í™•ë¥ ì„ ê³„ì‚°í•˜ì.

```python
logp = sequence_logprob(model, output_greedy, input_len=len(input_ids[0]))
print(tokenizer.decode(output_greedy[0]))
print(f"\në¡œê·¸ í™•ë¥ : {logp:.2f}")
```

ì´ê²ƒì„ beam searchë¡œ ìƒì„±í•œ ì‹œí€€ìŠ¤ì™€ ë¹„êµí•´ë³´ì

```python
output_beam = model.generate(input_ids, max_length=max_length, num_beams=5,
                             do_sample=False)
logp = sequence_logprob(model, output_beam, input_len=len(input_ids[0]))
print(tokenizer.decode(output_beam[0]))
print(f"\në¡œê·¸ í™•ë¥ : {logp:.2f}")
```

- í™•ë¥ ì´ ë†’ì•„ì¡Œì§€ë§Œ ì•„ì§ë„ ë°˜ë³µë˜ëŠ” ë¬¸ì œê°€ ìˆë‹¤. 
- ëŒ€ì±…: `no_repeat_ngram_size` íŒ¨ë„í‹° ê°’ì„ ì§€ì •í•œë‹¤. --> ì´ì „ì— ë‚˜ì˜¨ n-gramì„ ë˜ ìƒì„±í•˜ë©´ ë‹¤ìŒ í† í° í™•ë¥ ì´ 0ì´ ëœë‹¤.- 

```python
output_beam = model.generate(input_ids, max_length=max_length, num_beams=5,
                             do_sample=False, no_repeat_ngram_size=2)
logp = sequence_logprob(model, output_beam, input_len=len(input_ids[0]))
print(tokenizer.decode(output_beam[0]))
print(f"\në¡œê·¸ í™•ë¥ : {logp:.2f}")
```

## 5.4 ìƒ˜í”Œë§ ë°©ë²•

ê°€ì¥ ì†ì‰¬ìš´ ìƒ˜í”Œë§ ë°©ë²•ì€ ê° íƒ€ì„ìŠ¤í…ì— ëª¨ë¸ì´ ì¶œë ¥í•œ ì „ì²´ ì–´íœ˜ì‚¬ì „ì— ëŒ€í•œ ë¶„í¬ì—ì„œ ëœë¤í•˜ê²Œ í‘œì§‘í•˜ëŠ” ê²ƒì´ë‹¤.<br>
<br>
$$P(y_t=w_i | y_{<t}, x) = \mathrm{softmax}(z_{t, i}) = \frac{\mathrm{exp}(z_{t, i})}{\sum_{j=1}^{|V|}\mathrm{exp}(z_{t, j})}$$

- Softmax í•¨ìˆ˜ì— Temperature íŒŒë¼ë¯¸í„° Të¥¼ ì ìš©í•˜ë©´ ì¶œë ¥ì˜ ë‹¤ì–‘ì„±ì´ ì‰½ê²Œ ì œì–´ëœë‹¤.
- T ê°’ì„ ë°”ê¾¸ë©´ í™•ë¥ ë¶„í¬ì˜ í˜•íƒœê°€ ì œì–´ëœë‹¤.<br>
<br>
$$P(y_t=w_i | y_{<t}, x) = \frac{\mathrm{exp}(z_{t, i}/T)}{\sum_{j=1}^{|V|}\mathrm{exp}(z_{t, j}/T)}$$

```python
#ì„¸ ê°œì˜ ì˜¨ë„ì—ì„œ ëœë¤í•˜ê²Œ ìƒì„±í•œ í† í° í™•ë¥ ì˜ ë¶„í¬
import matplotlib.pyplot as plt
import numpy as np

def softmax(logits, T=1):
    e_x = np.exp(logits / T)
    return e_x / e_x.sum()

logits = np.exp(np.random.random(1000))
sorted_logits = np.sort(logits)[::-1]
x = np.arange(1000)

for T in [0.5, 1.0, 2.0]:
    plt.step(x, softmax(sorted_logits, T), label=f"T={T}")
plt.legend(loc="best")
plt.xlabel("Sorted token probabilities")
plt.ylabel("Probability")
plt.show()
```

- <span style="color:firebrick">**Top-k sampling** </span>: í™•ë¥ ì´ ë†’ì€ kê°œì˜ í† í°ë“¤ ì¤‘ì—ì„œë§Œ ìƒ˜í”Œë§í•´ì„œ ë‹¤ìŒ í† í°ì„ ì˜ˆì¸¡

```python
input_txt = """In a shocking finding, scientist discovered \
a herd of unicorns living in a remote, previously unexplored \
valley, in the Andes Mountains. Even more surprising to the \
researchers was the fact that the unicorns spoke perfect English.\n\n
"""
input_ids = tokenizer(input_txt, return_tensors="pt")["input_ids"].to(device)
```

```python
import torch.nn.functional as F

with torch.no_grad():
    output = model(input_ids=input_ids)
    next_token_logits = output.logits[:, -1, :]
    probs = F.softmax(next_token_logits, dim=-1).detach().cpu().numpy()
```

```python
#ë‹¤ìŒ í† í° ì˜ˆì¸¡ì˜ í™•ë¥  ë¶„í¬(ì™¼ìª½)ê³¼ ë‚´ë¦¼ì°¨ìˆœìœ¼ë¡œ ì •ë ¬ëœ í† í° í™•ë¥ ì˜ ëˆ„ì  ë¶„í¬
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(1, 2, figsize=(10, 3.5))

axes[0].hist(probs[0], bins=np.logspace(-10, -1, 100), color="C0", edgecolor="C0")
axes[0].set_xscale("log")
axes[0].set_yscale("log")
axes[0].set_title("Probability distribution")
axes[0].set_xlabel("Probability")
axes[0].set_ylabel("Count")
#axes[0].grid(which="major")

axes[1].plot(np.cumsum(np.sort(probs[0])[::-1]), color="black")
axes[1].set_xlim([0, 10000])
axes[1].set_ylim([0.75, 1.01])
axes[1].set_title("Cumulative probability")
axes[1].set_ylabel("Probability")
axes[1].set_xlabel("Token (descending probability)")
#axes[1].grid(which="major")
axes[1].minorticks_on()
#axes[1].grid(which='minor', linewidth='0.5')
top_k_label = 'top-k threshold (k=2000)'
top_p_label = 'nucleus threshold (p=0.95)'
axes[1].vlines(x=2000, ymin=0, ymax=2, color='C0', label=top_k_label)
axes[1].hlines(y=0.95, xmin=0, xmax=10000, color='C1', label=top_p_label, linestyle='--')
axes[1].legend(loc='lower right')
plt.tight_layout()
```
